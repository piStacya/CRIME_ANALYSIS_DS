```python
#NR - absolute number
#P_HTHAB - Rate per 100.000 population size.
import pandas as pd
import numpy as np

df_nuts = pd.read_csv(
    "../data/cleanedDatasets/nan_rows/regionDatasets/police_recorded_offences_by_NUTS_3_region_new.tsv", sep="\t")

df_offence = pd.read_csv("../data/cleanedDatasets/nan_rows/police_recorded_offences_by_offence_category_new.tsv", sep="\t")

#loen sisse failid
#praegune plaan:
#proovin eraldada kuritööd kategooriate kaupa, kus iga kuritöö kohta on riigi nimi ja kahes osas mõõtmed
#proovin eraldada riigid, igas tabelis oleks üks riik ja selle kuritööd ning kuritöö arvud
# 2) Lisa veerg 'group', mis sisaldab ICCS esimest 6 märki (ICCS01, ICCS02, ...)
df = df_offence.copy(deep=True)
df["group"] = df["iccs"].str[:6]

# 3) Leia kõik grupid
groups = df["group"].unique()

print("Leitud ICCS grupid:", groups)
```

```python
import os
# 3) Loo kaust 'ICCS' kui seda pole
output_dir = "../data/cleanedDatasets/nan_rows/ICCS/"
os.makedirs(output_dir, exist_ok=True)
# 4) Ekspordi iga grupp eraldi faili
for g in df["group"].unique():
    subset = df[df["group"] == g]

    # Failide täielik tee: ICCS/ICCS01.tsv, ICCS/ICCS02.tsv, ...
    output_path = os.path.join(output_dir, f"{g}.tsv")

    subset.to_csv(output_path, sep="\t", index=False)
    print(f"Loodud: {output_path}  (ridasid: {len(subset)})")
```

```python
countries_dir = "../data/cleanedDatasets/nan_rows/countries/"
os.makedirs(countries_dir, exist_ok=True)
for country in df["geo"].unique():
    subset2 = df[df["geo"] == country]
    output_path2 = os.path.join(countries_dir, f"{country}.tsv")
    subset2.to_csv(output_path2, sep="\t", index=False)
    print(f"Loodud: {output_path2}  (ridasid: {len(subset2)})")
```
